```python
# Loop through each column in the DataFrame
for column in df.columns:
    # Print the name of the current column
    print('columns ..... \n', column)
    # Find unique values in the current column
    unique_values = df[column].unique()
    # Print the unique values
    print(unique_values)
```

```python
# Define the columns to check for wrong values
wrong_values_colums = ["Age", "Educational Qualifications"]
```

```python
# Define the selected column for analysis
selceted_column = "Age"
```

```python
# Replace '?' values in the selected column with 0
df.loc[df[selceted_column] == '?', selceted_column] = 0
```

```python
# Loop through each column in the DataFrame
for column in df.columns:
    # Print the name of the current column
    print('columns ..... \n', column)
    # Find unique values in the current column
    unique_values = df[column].unique()
    # Print the unique values
    print(unique_values)
```

```python
# Loop through each column in the list
for column in wrong_values_colums:
    # Calculate the mode value of the current column
    mode_value = df[column].mode().iloc[0]
    print('mode value: >>>>>>>>>', mode_value)
    # Replace '?' values in the current column with the mode value
    df.loc[df[column] == '?', column] = mode_value
    
```

```python
# Loop through each column in the DataFrame
for column in df.columns:
    # Print the name of the current column
    print('columns ..... \n', column)
    # Find unique values in the current column
    unique_values = df[column].unique()
    # Print the unique values
    print(unique_values)
```
